### Moderator Priority Survey Results
* Used for calculating weights of labels

```python
import pandas as pd
import numpy as np
```

```python
survey = pd.read_csv('Toxicity Severity (Responses) - Form Responses 2.csv')
# Removed timestamps and email addresses
```

```python
data = survey.iloc[:, 2:7]
data.head()
```

```python
data.columns = ['toxic', 'obscene', 'threat', 'insult', 'identity_hate']
```

```python
data.head()
```

```python
data
```

```python
data = data.replace('5 (great concern, should be given priority)', '5')
data = data.replace('1 (little to no concern)', '1')
```

```python
data
```

```python
data = data.astype('int32')
```

```python
data.head()
```

```python
data['toxic'][0]
```

### Get the percentage for each label

```python
percent_data = pd.DataFrame()
```

```python
# For the first label 'toxic'
toxic_1 = len(data['toxic'][data['toxic'] == 1]) / len(data)
toxic_2 = len(data['toxic'][data['toxic'] == 2]) / len(data)
toxic_3 = len(data['toxic'][data['toxic'] == 3]) / len(data)
toxic_4 = len(data['toxic'][data['toxic'] == 4]) / len(data)
toxic_5 = len(data['toxic'][data['toxic'] == 5]) / len(data)

toxic_arr = np.array([toxic_1, toxic_2, toxic_3, toxic_4, toxic_5])
toxic_arr
```

```python
# For the second label 'obscene'
obscene_1 = len(data['obscene'][data['obscene'] == 1]) / len(data)
obscene_2 = len(data['obscene'][data['obscene'] == 2]) / len(data)
obscene_3 = len(data['obscene'][data['obscene'] == 3]) / len(data)
obscene_4 = len(data['obscene'][data['obscene'] == 4]) / len(data)
obscene_5 = len(data['obscene'][data['obscene'] == 5]) / len(data)

obscene_arr = np.array([obscene_1, obscene_2, obscene_3, obscene_4, obscene_5])
obscene_arr
```

```python
# For the third label 'threat'
threat_1 = len(data['threat'][data['threat'] == 1]) / len(data)
threat_2 = len(data['threat'][data['threat'] == 2]) / len(data)
threat_3 = len(data['threat'][data['threat'] == 3]) / len(data)
threat_4 = len(data['threat'][data['threat'] == 4]) / len(data)
threat_5 = len(data['threat'][data['threat'] == 5]) / len(data)

threat_arr = np.array([threat_1, threat_2, threat_3, threat_4, threat_5])
threat_arr
```

```python
# For the fourth label 'insult'
insult_1 = len(data['insult'][data['insult'] == 1]) / len(data)
insult_2 = len(data['insult'][data['insult'] == 2]) / len(data)
insult_3 = len(data['insult'][data['insult'] == 3]) / len(data)
insult_4 = len(data['insult'][data['insult'] == 4]) / len(data)
insult_5 = len(data['insult'][data['insult'] == 5]) / len(data)

insult_arr = np.array([insult_1, insult_2, insult_3, insult_4, insult_5])
insult_arr
```

```python
# For the fifth label 'identity_hate'
identity_hate_1 = len(data['identity_hate'][data['identity_hate'] == 1]) / len(data)
identity_hate_2 = len(data['identity_hate'][data['identity_hate'] == 2]) / len(data)
identity_hate_3 = len(data['identity_hate'][data['identity_hate'] == 3]) / len(data)
identity_hate_4 = len(data['identity_hate'][data['identity_hate'] == 4]) / len(data)
identity_hate_5 = len(data['identity_hate'][data['identity_hate'] == 5]) / len(data)

identity_hate_arr = np.array([identity_hate_1, identity_hate_2, identity_hate_3, identity_hate_4, identity_hate_5])
identity_hate_arr
```

```python
percent_data['Toxic Percentage'] = toxic_arr
percent_data['Obscene Percentage'] = obscene_arr
percent_data['Threat Percentage'] = threat_arr
percent_data['Insult Percentage'] = insult_arr
percent_data['Identity_Hate Percentage'] = identity_hate_arr
```

```python
percent_data = percent_data.rename({0:1, 1:2, 2:3, 3:4, 4:5}, axis='index')
```

```python
percent_data
```

```python
percent_data.to_csv('percent_data.csv')
```
